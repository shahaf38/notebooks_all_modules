# Acoustic features

## Load data

```python
# load the data with DF
query = ''' SELECT * FROM acoustic_features'''
acoustic_features_df= pd.read_sql_query(query, con=bdd)

acoustic_features_df.head()
```

## Mean by year

```python
# input df
# output df with a index as Date index
# get the date and transform it into Date index

dates = acoustic_features_df['date']
date_index = pd.to_datetime(dates, format='mixed')
acoustic_features_df = acoustic_features_df.set_index(date_index)
```

```python
#Creating a mask and take only the relevant columns
keep_columns = (acoustic_features_df.dtypes == float)
tmp = acoustic_features_df.loc[:, keep_columns].resample('Y').mean().dropna(axis=0)
```

```python
tmp.acousticness.plot()
```

```python
# Iterate through each column in tmp DataFrame
for column in tmp.columns:
    plt.figure()  # Create a new figure for each plot
    tmp[column].plot(title=f'{column} Over Time', xlabel='Date', ylabel=column)
    plt.show()  # Display the plot
```

we can see in all the graphs that the 1960 influenced music in all areas. Not surprising because this is the period of the Vietnam War.
